## 1. Cargamos dataset

```python
import pandas as pd

pd.options.display.max_colwidth = 10000
df = pd.read_csv("Python_Máster/datasets/villagers.csv")
df = df.reset_index(drop=True)
df.head(5)
```

## 2. Convertimos la fecha en formato date

```python
# hacemos split

df[['Month','Day']] = df.Birthday.str.split("/",expand=True,)

#creamos año

df['Year'] = 2001

# convertimos a strings y creamos fecha

df["Cumple"] = df["Day"].astype(str) + "-" + df["Month"].astype(str) + "-" + df["Year"].astype(str) 

# drop columnas 


# cambiamos a formato fecha fecha

df["Cumple"] =  pd.to_datetime(df['Cumple'], format='%d-%m-%Y')

df.drop(["Birthday", 'Month', 'Day', 'Year'], axis=1, inplace=True)

df
```

## 3. Cargamos los módulos necesarios

```python
import requests
from IPython.display import display
from PIL import Image
```

## 4. ¡A ver qué personaje sale!

```python
year = "2001"
print("¿En qué mes cumples años? Por favor, inserta el número")
mes = str(input())
print("¿En qué día cumples años?")
day = str(input())
date = day  + "-"+ mes +"-"+ year
try: 
    date =  pd.to_datetime(date, format='%d-%m-%Y')
except ValueError:
    print("te has confundido")
date
pd.options.display.max_colwidth = 10000
a = df.loc[(df["Cumple"] == date)]
img_url = list(a["Photo Image"])
name = list(a["Name"])
house_url = list(a["House Image"])
persona = list(a["Personality"])
frase = list(a["Catchphrase"])
cancion = list(a["Favorite Song"])
for i, j, l, m, q, k in zip(img_url, name, persona, frase, cancion, house_url):    # lista comprensiva
    img = Image.open(requests.get(i, stream = True).raw)
    print("Compartes cumple con {}".format(j))
    display(img)
    print("Su personalidad es: {}".format(l))
    print("Siempre dice {}".format(m))
    print("Su canción favorita es {}".format(q))
    img0 = Image.open(requests.get(k, stream = True).raw)
    print("Y vive en esta casa")
    display(img0)
```
